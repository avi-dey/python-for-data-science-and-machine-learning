## Creating a Figure

The main idea in using the more formal Object Oriented method is to create figure objects and then just call methods or attributes off of that object.\
This approach is nicer when dealing with a canvas that has multiple plots on it. 

```python
# Creates blank canvas
fig = plt.figure()
# default size is 432x288
# you can pass parameters to change it
```

```python
# shift + tab
fig = plt.figure(figsize=(12,8)) # 12 inches x 8 inches
# 12 inch is a little more than 30cm
# 6 inch is a lil more than 15cm
# when you save the plot as jpg/png/etc. you will get the acutal size
```

**NOTE: ALL THE COMMANDS NEED TO GO IN THE SAME CELL!**

To begin we create a figure instance. Then we can add axes to that figure:

```python
# Create Figure (empty canvas)
fig = plt.figure()

# Add set of axes to figure
axes = fig.add_axes([0, 0, 1, 1]) # left, bottom, width, height (range 0 to 1)

# Plot on that set of axes
axes.plot(x, y)

plt.show()
```

```python
# Create Figure (empty canvas)
fig = plt.figure()

# Add set of axes to figure
axes = fig.add_axes([0, 0, 1, 1]) # left, bottom, width, height (range 0 to 1)

# Plot on that set of axes
axes.plot(a, b)

plt.show()
```

### Adding another set of axes to the Figure

So far we've only seen one set of axes on this figure object, but we can keep adding new axes on to it at any location and size we want. We can then plot on that new set of axes.

```python
type(fig)
```

Code is a little more complicated, but the advantage is that we now have full control of where the plot axes are placed, and we can easily add more than one axis to the figure. Note how we're plotting a,b twice here

```python
axes3 = fig.add_axes([0.9,0.9,1,1])
m = np.linspace(0,11,12)
n = np.arange(0,12)
axes3.plot(m,n)
axes3.set_title("hello world")
# what is happening to this?

# Creates blank canvas
fig = plt.figure()

axes1 = fig.add_axes([0, 0, 0.9, 0.9]) # Large figure
axes2 = fig.add_axes([0.2, 0.2, 0.4, 0.4]) # Smaller figure

# Larger Figure Axes 1
axes1.plot(a, b)

# Use set_ to add to the axes figure
axes1.set_xlabel('X Label')
axes1.set_ylabel('Y Label')
axes1.set_title('Big Figure')

# Insert Figure Axes 2
axes2.plot(a,b)
axes2.set_title('Small Figure');

```

Let's move the small figure and edit its parameters.

```python
# Creates blank canvas
fig = plt.figure()

axes1 = fig.add_axes([0, 0, 1, 1]) # Large figure
axes2 = fig.add_axes([0.2, 0.5, 0.25, 0.25]) # Smaller figure

# Larger Figure Axes 1
axes1.plot(a, b)

# Use set_ to add to the axes figure
axes1.set_xlabel('X Label')
axes1.set_ylabel('Y Label')
axes1.set_title('Big Figure')

# Insert Figure Axes 2
axes2.plot(a,b)
axes2.set_xlim(8,10)
axes2.set_ylim(4000,10000)
axes2.set_xlabel('X')
axes2.set_ylabel('Y')
axes2.set_title('Zoomed In');
```

You can add as many axes on to the same figure as you want, **even outside of the main figure** if the length and width correspond to this.

```python
# Creates blank canvas
fig = plt.figure()

axes1 = fig.add_axes([0, 0, 1, 1]) # Full figure
axes2 = fig.add_axes([0.2, 0.5, 0.25, 0.25]) # Smaller figure
axes3 = fig.add_axes([1, 1, 0.25, 0.25]) # Starts at top right corner!

# Larger Figure Axes 1
axes1.plot(a, b)

# Use set_ to add to the axes figure
axes1.set_xlabel('X Label')
axes1.set_ylabel('Y Label')
axes1.set_title('Big Figure')

# Insert Figure Axes 2
axes2.plot(a,b)
axes2.set_xlim(8,10)
axes2.set_ylim(4000,10000)
axes2.set_xlabel('X')
axes2.set_ylabel('Y')
axes2.set_title('Zoomed In');

# Insert Figure Axes 3
axes3.plot(a,b)
```

## Figure Parameters
earlier you were not passing any parameters to figure()

```python
# Creates blank canvas
fig = plt.figure(figsize=(12,8),dpi=100) # dpi = dots per inch, *** more dpi takes up more RAM ***

axes1 = fig.add_axes([0, 0, 1, 1])

axes1.plot(a,b)
```

## Exporting/Saving a Figure

```python
fig = plt.figure()

axes1 = fig.add_axes([0, 0, 1, 1])

axes1.plot(a,b)
axes1.set_xlabel('X')

# setting the bounding box (bbox="tight") to tight would show us the x and y axes in the image which we save
# bbox_inches ='tight' automatically makes sure the bounding box is correct
fig.savefig('figure.png', bbox_inches='tight')
```
