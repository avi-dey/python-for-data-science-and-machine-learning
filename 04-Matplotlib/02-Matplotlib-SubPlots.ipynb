# plt.subplots()

**NOTE: Make sure you put the commands all together in the same cell as we do in this notebook and video!**

The plt.subplots() object will act as a more automatic axis manager. This makes it much easier to show multiple plots side by side.

Note how we use tuple unpacking to grba both the Figure object and a numpy array of axes:

```python
# Use similar to plt.figure() except use tuple unpacking to grab fig and axes
fig, axes = plt.subplots()

# Now use the axes object to add stuff to plot
axes.plot(x, y, 'r')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('title'); #; hides Out[]
```

## Plotting on axes objects

Just as before, we simple .plot() on the axes objects, and we can also use the .set_ methods on each axes.

Let's explore this, make sure this is all in the same cell:

```python
fig,axes = plt.subplots(nrows=1,ncols=3)

# we can use the for loop on column vectors(1 by n) and row vector(n by 1) 
for axe in axes:
    axe.plot(x,y)
    
# fix the overlap with
# fig.tight_layout()
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)

# just how you would select an array, as subplots are nothing but numpy array
axes[0].plot(a,b)
axes[1].plot(x,y)
```

```python
# NOTE! This returns 2 dimensional array
fig,axes = plt.subplots(nrows=2,ncols=2)

# just how you would select a 2 dimensional array, as subplots are nothing but numpy array(here it is numpy 2d array)
axes[0][0].plot(a,b)
axes[1][0].plot(x,y)
```

A common issue with matplolib is overlapping subplots or figures. We ca use **fig.tight_layout()** or **plt.tight_layout()** method, which automatically adjusts the positions of the axes on the figure canvas so that there is no overlapping content:

```python
# NOTE! This returns 2 dimensional array
fig,axes = plt.subplots(nrows=2,ncols=2)

axes[0][1].plot(a,b)
axes[1][1].plot(x,y)  

# use tight_layout() to fix the overlapping issue between two subplots
plt.tight_layout()
```

## Parameters on subplots()

Recall we have both the Figure object and the axes. Meaning we can edit properties at both levels.

```python
fig,axes = plt.subplots(nrows=2,ncols=2,figsize=(12,8))

# SET YOUR AXES PARAMETERS FIRST

# Parameters at the axes level
# 0 0
axes[0][0].plot(a,b)
axes[0][0].set_title('0 0 Title')

# 1 1
axes[1][1].plot(x,y)
axes[1][1].set_title('1 1 Title')
axes[1][1].set_xlabel('1 1 X Label')

# 0 1
axes[0][1].plot(y,x)
axes[0][1].set_ylabel("0 1 Y Label")

# 1 0
axes[1][0].plot(b,a)
axes[1][0].set_ylabel("1 0 Y Label")
# THEN SET OVERALL FIGURE PARAMETERS

# Parameters at the Figure level
# Super Title
fig.suptitle("Figure Level",fontsize=16)


plt.show()
```

### Manual spacing on subplots()

If you don't like what fig.tight_layout() or plt.tight_layout() offers by default, you can set your manual custom spacing. Use .subplots_adjust to adjust spacing manually to what you desire.

Full Details Here: https://matplotlib.org/3.2.2/api/_as_gen/matplotlib.pyplot.subplots_adjust.html

Example from link:

* left = 0.125  # the left side of the subplots of the figure
* right = 0.9   # the right side of the subplots of the figure
* bottom = 0.1  # the bottom of the subplots of the figure
* top = 0.9     # the top of the subplots of the figure
* wspace = 0.2  # the amount of width reserved for space between subplots,
              # expressed as a fraction of the average axis width
* hspace = 0.2  # the amount of height reserved for space between subplots,
              # expressed as a fraction of the average axis height
              
<img src="subplots_adjust__idea_of_sizes.png">

### 1 means an entire subplot can fit into that space

```python
fig,axes = plt.subplots(nrows=2,ncols=2,figsize=(12,8))

# Parameters at the axes level
axes[0][0].plot(a,b)
axes[1][1].plot(x,y)
axes[0][1].plot(y,x)
axes[1][0].plot(b,a)

# Use left,right,top, bottom to stretch subplots
# Use wspace,hspace to add spacing between subplots
# 1 means an entire subplot can fit into the space
# 0.9 means 90% of the entire subplot can fit into that space
# and so on
fig.subplots_adjust(left=None,
    bottom=None,
    right=None,
    top=None,
    wspace=0.9,
    hspace=0.1,)

plt.show()
```

## Exporting/Saving plt.subplots()

```python
# NOTE! This returns 2 dimensional array
## **fig_kw is accepted as parameters, means all those parameters used in figure() parameters are accepted here
## that is the reason parameters like "dpi" is also accepted
fig,axes = plt.subplots(nrows=2,ncols=2,figsize=(12,8),dpi= 200)

axes[0][0].plot(a,b)
axes[1][1].plot(x,y)
axes[0][1].plot(y,x)
axes[1][0].plot(b,a)

fig.savefig('subplots.png',bbox_inches='tight') # bounded box = tight so that our image also saves the x and y axes labelling and label names

plt.show()
```
