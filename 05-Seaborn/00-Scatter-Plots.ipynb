# Scatter Plots

Scatter plots can show how different features are related to one another, the main theme between all relational plot types is they display how features are interconnected to each other. There are many different types of plots that can be used to show this, so let's explore the scatterplot() as well as general seaborn parameters applicable to other plot types.

----
## Data

We'll use some generated data from: http://roycekimmons.com/tools/generated_data



```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv("dm_office_sales.csv")
```

```python
df.head()
```

```python
df.info()
```

-----

# Scatterplot

```python
sns.scatterplot(x='salary',y='sales',data=df)
```

## Connecting to Figure in Matplotlib

#### Seaborn uses matplotlib in the background/backend that means you can use matplotlib commands on seaborn objects directly

**Note how matplotlib is still connected to seaborn underneath (even without importing matplotlib.pyplot), since seaborn itself is directly making a Figure call with matplotlib at the backend. We can import matplotlib.pyplot and make calls to directly effect the seaborn figure.**

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(12,4), dpi= 200) # matplot lib figure() directly affects the seaborn scatterplot axes
sns.scatterplot(x='salary',y='sales',data=df)
```

## Seaborn Parameters

The hue and palette parameters are commonly available around many plot calls in seaborn.

### hue
We can color points based of a _"categorical feature -- distinct like name/color/shape/some-grouping-name/etc -- like discrete random variable"_ or _"continuous feature -- ages/height/weight/salary/temperature/prices/etc -- like continuous random variable"_

it's like plotting 3 featues on a 2d graph, the 3rd feature is given with color

#### 1 Color points based off a "categorical feature" in the DataFrame:

```python
plt.figure(figsize=(12,4))
sns.scatterplot(x='salary',y='sales',data=df, hue='division')
```

#### 2 Coloring based on "continuous feature" in the DF:

```python
plt.figure(figsize=(12,4), dpi=200)
sns.scatterplot(x='salary',y='sales',data=df,hue='work experience')
```

apart from the parameters "x, y, data and hue" everything else is for styling
# Styling

**Choosing a color palette from Matplotlib's cmap:https://matplotlib.org/stable/tutorials/colors/colormaps.html**
you can choose a color palette from here to style your hue color

```python
# example 1
plt.figure(figsize=(12,3))
sns.scatterplot(x='salary',y='sales',data=df,hue='work experience',palette='viridis')
```

```python
# example 2 
plt.figure(figsize=(12,3))
sns.scatterplot(x='salary',y='sales',data=df,hue='work experience',palette='Dark2')
```

## Scatterplot  Parameters

These parameters are more specific to the scatterplot() call

### size

Allows you to size dots based on another column in DF 

```python
plt.figure(figsize=(12,3))
sns.scatterplot(x='salary', y='sales', data=df, size='work experience');
# check the size of the dots now, accoring to the work experience
# least work exp - smallest dots, more work exp - bigger dots
```

## s
**Use *s=* if you want to change the marker size to be some uniform integer value** <br>
make the marker bigger or smaller

```python
plt.figure(figsize=(12,3))
sns.scatterplot(x='salary',y='sales',data=df,s=200)
```

```python
# adding linewidth i.e. borderwidth of the points
plt.figure(figsize=(12,3))
sns.scatterplot(x='salary',y='sales',data=df,s=200,linewidth=3)
```

## alpha
1 - completely opaque <br>
0 - completely transparent <br>
[0,1] is the range 

* linewidth is the borderwidth of the dots

```python
plt.figure(figsize=(12,3))
sns.scatterplot(x='salary',y='sales',data=df,s=200,alpha=0.2) # default linewidth = 1
```

```python
plt.figure(figsize=(12,3))
sns.scatterplot(x='salary',y='sales',data=df,s=200,alpha=0.2, linewidth = 0.5)
```

```python
plt.figure(figsize=(12,3))
sns.scatterplot(x='salary',y='sales',data=df,s=200,alpha=0.2, linewidth = 0.1)
```

### style
Sets the shape of dots <br>
Automatically choose styles based on another categorical feature in the dataset. Optionally use the **markers=** parameter to pass a list of marker choices based off matplotlib, for example: ['*','+','o']

```python
plt.figure(figsize=(12,3))
sns.scatterplot(x='salary', y='sales', data=df, style='level of education')
```

```python
# combining hue<coloring> and style<shapes> parameters to get both coloring and shape

plt.figure(figsize=(12,3))
# Sometimes its nice to do BOTH hue and style off the same column
sns.scatterplot(x='salary',y='sales',data=df,style='level of education',hue='level of education',s=100)
```

## Exporting/Saving a Seaborn Figure

```python
plt.figure(figsize=(12,3))
sns.scatterplot(x='salary',y='sales',data=df,style='level of education',hue='level of education',s=100)

# Call savefig in the same cell
plt.savefig('example_scatter.jpg')
```
