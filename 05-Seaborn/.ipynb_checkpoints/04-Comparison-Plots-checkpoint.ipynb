# Comparison Plots with pairplot() and jointplot()
1. [jointplot](#joint)
2. [pairplot](#pair)

## Imports

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## The Data

```python
df = pd.read_csv("StudentsPerformance.csv")
```

```python
df.head()
```

<a id="joint"></a>
# jointplot

```python
plt.figure(figsize=(10,5))
sns.jointplot(x='math score',y='reading score',data=df)
```

```python
sns.jointplot(x='math score',y='reading score',data=df, alpha=0.2)
```

```python
sns.jointplot(x='math score',y='reading score',data=df,kind='hex')
```

```python
sns.jointplot(x='math score',y='reading score',data=df,kind='hist')
```

```python
sns.jointplot(x='math score',y='reading score',data=df,kind='kde')
```

```python
sns.jointplot(x='math score',y='reading score',data=df,kind='kde',fill=True)
```

```python
 # cmap sets color of fill from Matplot colormap options
sns.jointplot(x='math score',y='reading score',data=df,kind='kde',fill=True,cmap="Greens")
```

```python
# more parameters
plt.figure(figsize=(2,1))
sns.jointplot(x='math score',y='reading score', hue="gender",data=df)
```

<a id="pair"></a>
# pairplot

```python
sns.pairplot(df)
```

```python
# adding hue, automatically converts diagonals to kde plot
sns.pairplot(df,hue='gender',palette='viridis') 
```

```python
# if you want to keep diagonals as histograms
sns.pairplot(data=df, hue="gender",diag_kind="hist")
```

```python
# since, upper and lower triangle of the plots are same, cut one triangle out with "corner" parameter
sns.pairplot(df,hue='gender',palette='viridis',corner=True)
```
