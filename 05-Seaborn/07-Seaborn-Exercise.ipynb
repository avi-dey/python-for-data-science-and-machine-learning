### TASK: Recreate the Scatter Plot shown below

**The scatterplot attempts to show the relationship between the days employed versus the age of the person (DAYS_BIRTH) for people who were not unemployed. Note, to reproduce this chart you must remove unemployed people from the dataset first. Also note the sign of the axis, they are both transformed to be positive. Finally, feel free to adjust the *alpha* and *linewidth* parameters in the scatterplot since there are so many points stacked on top of each other.** 

<img src="task_one.jpg">


```python
import warnings
warnings.simplefilter("ignore")
```

```python
# CODE HERE TO RECREATE THE PLOT SHOWN ABOVE
plt.figure(figsize=(12,8))

# new dataframe free of unemployed people
# remove unemployed people (negative values tells employment days)
employed = df[df["DAYS_EMPLOYED"] < 0]

# converting negative values to positive data
employed["DAYS_EMPLOYED"] = -1 * employed["DAYS_EMPLOYED"]
employed["DAYS_BIRTH"] = -1 * employed["DAYS_BIRTH"]

# plotting
sns.scatterplot(x='DAYS_BIRTH', y="DAYS_EMPLOYED",data=employed, alpha = 0.01) # linewidth ranges from 1 and above
```

```python
# CODE HERE TO RECREATE THE PLOT SHOWN ABOVE

# converting the negative to positive
new_df = df[["DAYS_BIRTH"]] 
new_df["DAYS_BIRTH"] = -1 * new_df["DAYS_BIRTH"]
new_df["YEARS"] = new_df["DAYS_BIRTH"]/365 # getting years

plt.figure(figsize=(8,4))
# plotting
sns.histplot(data=new_df, x="YEARS",color="red",
            edgecolor="black", linewidth=3,
            bins=45,
            alpha=0.5)

plt.xlabel("Age in Years")
plt.tight_layout()
```

### TASK: Recreate the Categorical Plot shown below:

<img src='catplot_solution.png'>

**This plot shows information only for the *bottom half* of income earners in the data set. It shows the boxplots for each category of NAME_FAMILY_STATUS column for displaying their distribution of their total income. Note: You will need to adjust or only take part of the dataframe *before* recreating this plot. You may want to explore the *order* parameter to get the xticks in the exact order shown here**

```python
# CODE HERE
plt.figure(figsize=(12,5))

# creating the bottom half income df
bottom_half_income = df.nsmallest(n=int(0.5 * len(df)),
                                 columns="AMT_INCOME_TOTAL")
# PLOT
sns.boxplot(x="NAME_FAMILY_STATUS", y="AMT_INCOME_TOTAL", hue="FLAG_OWN_REALTY",
            data=bottom_half_income)

plt.legend(bbox_to_anchor=(1.05, 1), borderaxespad=0., title='FLAG_OWN_REALTY')
plt.title("Income per family for bottom half of the Earners")
```

```python
## only one triangle
plt.figure(figsize=(12,5))

import numpy as np

df_corr = df.drop("FLAG_MOBIL",axis=1).corr(numeric_only=True) # modified df
matrix = np.triu(df_corr) # selecting upper triangle, we will mask this

sns.heatmap(data=df_corr, mask=matrix,
           annot=True, cmap="viridis") # masks the upper triangle
```
