### apply with lambda

In Python, a **lambda function** (also known as an anonymous function) is a small, one-line function that can be defined without a name. <br>
Syntax: `lambda arguments: expression` <br>
Here, `arguments` are the input arguments to the function and `expression` <span style="color:lime;"> is a single expression that is evaluated and returned </span> as the result of the function. The lambda keyword is used to define the function, and the function is not assigned to a variable. (source: chatGPT)

```python
# simple function
def simple(num):
    return num*2
```

```python
lambda num: num*2 # lambda function
```

Either of these two methods gives the same results. <br> 
`np.vectorize` method is a lot simpler BUT `np.vectorize` is also computationally very very fast.
So the obvious question here is what the heck is `np.vectorize` doing, that it is so faster (we'll prove it later with `timeit` module.) <br>

And it's especially confusing because if you actually look up the of [np.vectorize docs](https://numpy.org/doc/stable/reference/generated/numpy.vectorize.html) it is stated that "The vectorize function is provided primarily for <span style="color:green;">convenience, not for performance."</span> <br>

So it's confusing because the numpy documentation itself says this is for convenience, not for performance. Then why does it performs faster? (we'll prove later)<br>
So what the heck is this actually doing and what's going on here? <br>

Essentially what np.vectorize does and there's a nice [stackoverflow post](https://stackoverflow.com/questions/3379301/using-numpy-vectorize-on-functions-that-return-vectors) where the top answer by "unutbu" explains what is happening here.
He's says it is a convenience function and it does not make your code run any faster. 
So the main thing that's happening here is that the
purpose of `np.vectorize`, is to transform functions which are not numpy aware. <br>
And that's exactly what we have here.<br>
If we take a look back at this original `quality()` function, it is not aware that we're going to be passing in and broadcasting this to what is essentially a numpy array within the pandas series objects. <br>
So recall every pandas column in a data frame is infact a panda series, which in turn is holding a numpy array. <br>
And we know that if I broadcast specific numpy aware methods like dot(), mean() or something on an empty array, it can perform this in a very efficient manner. <br>

But right now I have what is essentially a standard Python function i.e. `quality()`.
It just accepts normal integers and floating point numbers.
It's not actually thinking of this as being cast or broadcasted to an numpy array. 
So if I actually transform this or vectorize this to be numpy aware, then it actually will get computationally more efficient. <br>

Not because of np.vectorize() so to speak, but just because I'm really transforming this from a normal function to then be aware of a numpy array so np.vectorize() won't actually improve performance if you're already dealing with something that's aware that it's taking in a numpy array.<br>

But I have the kind of the special case here, where this function is not aware that it's going to be taking in a numpy array.<br>

In fact, it's not even designed to really think of that that way.<br>

It's just taking in single values.<br>

So by vectorizing it, we actually are going to make this faster.<br>

```python
# shortcut to do the above with np.vectorize
# np.vectorize 
df['Tip Quality'] = np.vectorize(quality)(df['total_bill'], df['tip'])
```

```python
df.head()
```

So, which one is faster?

```python
import timeit 
  
# code snippet to be executed only once 
setup = '''
import numpy as np
import pandas as pd
df = pd.read_csv('tips.csv')
def quality(total_bill,tip):
    if tip/total_bill  > 0.25:
        return "Generous"
    else:
        return "Other"
'''
  
# code snippet whose execution time is to be measured 
stmt_one = ''' 
df['Tip Quality'] = df[['total_bill','tip']].apply(lambda df: quality(df['total_bill'],df['tip']),axis=1)
'''

stmt_two = '''
df['Tip Quality'] = np.vectorize(quality)(df['total_bill'], df['tip'])
'''
  
```

```python
timeit.timeit(setup = setup, 
                    stmt = stmt_one, 
                    number = 1000) 
```

```python
timeit.timeit(setup = setup, 
                    stmt = stmt_two, 
                    number = 1000) 
```

Wow! Vectorization is much faster! Keep **np.vectorize()** in mind for the future.

## between

left: A scalar value that defines the left boundary
right: A scalar value that defines the right boundary
inclusive: A Boolean value which is True by default. If False, it excludes the two passed arguments while checking.

```python
# df['total_bill'].between(10,20,inclusive=True)
# FutureWarning: Boolean inputs to the `inclusive` argument are deprecated in favour of `both` or `neither`.

df["total_bill"].between(10,20,inclusive="both") # both 10 and 20 are included i.e.[10,20] range
```

```python
df[df['total_bill'].between(10,20,inclusive="neither")] # (10,20) range i.e. 10 and 20 are excluded but everything b/w is taken
```
